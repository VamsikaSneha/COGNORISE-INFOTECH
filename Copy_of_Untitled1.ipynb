```python
# Age Distribution: A histogram to show the distribution of customer ages.

plt.figure(figsize=(8, 6))
plt.hist(Customertravel['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```

# RESULT

 The histogram provides a visual representation of how customers ages are distributed across the dataset, with the x-axis representing age ranges and the y-axis representing the count of customers in each range.

```python
# Frequent Flyer vs Non-Frequent Flyer: A bar chart to compare the number of frequent flyers against non-frequent flyers.

# Step 1: Count the occurrences of each category in the 'FrequentFlyer' column
flyer_counts = Customertravel['FrequentFlyer'].value_counts()

# Step 2: Plot the bar chart
plt.figure(figsize=(6, 4))
plt.bar(flyer_counts.index, flyer_counts.values, color=['skyblue', 'orange'])

# Step 3: Add titles and labels
plt.title('Frequent Flyer vs Non-Frequent Flyer')
plt.xlabel('Frequent Flyer Status')
plt.ylabel('Number of Customers')
plt.show()
```

# RESULT

The result of the bar chart comparing Frequent Flyers vs Non-Frequent Flyers would show two bars, one representing the number of customers who are Frequent Flyers (labeled "Yes") and the other representing the number of Non-Frequent Flyers (labeled "No").

The height of each bar corresponds to the count of customers in each category. This chart helps visualize how many customers belong to each group.

```python
# Income Class Distribution: A pie chart or bar chart showing the distribution of customers across different income classes.

import matplotlib.pyplot as plt

# Count the occurrences of each income class
income_class_counts = Customertravel['AnnualIncomeClass'].value_counts()

# Plot a pie chart
plt.figure(figsize=(6, 6))
plt.pie(income_class_counts.values, labels=income_class_counts.index, autopct='%1.1f%%', colors=['skyblue', 'orange', 'lightgreen'])
plt.title('Income Class Distribution of Customers')
plt.show()

# RESULT: It will display a circular chart divided into slices representing each income class.Each slice's size reflects the proportion of customers in that income class, and the percentages will be shown on each slice.
```

# RESULT

It will display a circular chart divided into slices representing each income class.

Each slice's size reflects the proportion of customers in that income class, and the percentages will be shown on each slice.

```python

# Services Opted: A boxplot to show the spread of services opted by customers, possibly grouped by income class.

# Set the plot size
plt.figure(figsize=(10, 6))

# Create the boxplot
sns.boxplot(x='AnnualIncomeClass', y='ServicesOpted', data=Customertravel, palette='Set3')

# Add title and labels
plt.title('Services Opted by Income Class')
plt.xlabel('Income Class')
plt.ylabel('Number of Services Opted')

# Show the plot
plt.show()
```

# **Explanation:**
Seaborn (sns): This is a library built on top of Matplotlib that makes it easier to create beautiful visualizations like boxplots.

boxplot(): This function creates a boxplot with:

* x='AnnualIncomeClass': Groups the data by income class.

* y='ServicesOpted': Plots the spread of services opted by customers.

* data=customer_travel_data: Uses the dataset.

* palette='Set3': A color palette to differentiate between the income classes.

This will produce a boxplot for each income class, showing how the number of services opted varies within each group. Youll be able to see the median, spread, and any outliers for each income class.

# RESULT

The result of the boxplot will show how the number of services opted by customers varies across different income classes. Each income class will have its own box, representing the spread of the data for that group.

Interpretation of the Boxplot:

Box: The central box represents the interquartile range (IQR), which contains the middle 50% of the data for each income class.

Median: A line inside the box represents the median (the middle value) of the number of services opted.

Whiskers: These extend from the box to show the range of the data (excluding outliers).

Outliers: Points outside the whiskers, representing values that are unusually high or low compared to the rest of the data.

# What I Learn

Spread of Services Opted: You'll see how the distribution of services opted varies between income classes (e.g., are certain income classes opting for more services on average?).

Median Comparison: The boxplot will help compare the median number of services opted for each income class.

Outliers: If there are customers in any income class opting for an unusually high or low number of services, they will appear as individual points.

```python
 # Account Synced to Social Media vs Booked Hotel: A stacked bar chart to visualize the relationship between social media syncing and hotel bookings.



# Step 1: Group the data by 'AccountSyncedToSocialMedia' and 'BookedHotelOrNot'
synced_hotel_data = Customertravel.groupby(['AccountSyncedToSocialMedia', 'BookedHotelOrNot']).size().unstack()

# Step 2: Plot the stacked bar chart
synced_hotel_data.plot(kind='bar', stacked=True, color=['skyblue', 'orange'], figsize=(8, 6))

# Step 3: Add titles and labels
plt.title('Account Synced to Social Media vs Booked Hotel')
plt.xlabel('Account Synced to Social Media')
plt.ylabel('Number of Customers')
plt.legend(title='Booked Hotel')
plt.show()
```

# Explanation

**Group the data:** We use groupby() to count the occurrences of each combination of "Account Synced to Social Media" and "Booked Hotel." Then, we use unstack() to reshape the data for easier plotting.

**Stacked Bar Chart**: The plot(kind='bar', stacked=True) function creates a stacked bar chart, where the bars are stacked by whether the customer booked a hotel.

**Colors:** We use two different colors for each booking status (booked or not booked).

**Legend:** The legend helps differentiate between customers who booked a hotel and those who did not.

**RESULT**

The x-axis will represent whether customers have synced their accounts to social media.

The bars will be divided into two segments: those who booked a hotel and those who didn't.

The y-axis will represent the number of customers.

This chart will help you understand whether syncing social media accounts correlates with hotel bookings.

```python
# Churn by Income Class: A grouped bar chart showing the churn rate (Target = 1) across different income classes.



# Step 1: Calculate churn rate for each income class
churn_rate = Customertravel.groupby('AnnualIncomeClass')['Target'].mean()

# Step 2: Prepare data for grouped bar chart
churn_rate = churn_rate.reset_index()

# Step 3: Plot the grouped bar chart
plt.figure(figsize=(8, 6))
plt.bar(churn_rate['AnnualIncomeClass'], churn_rate['Target'], color='skyblue')

# Step 4: Add titles and labels
plt.title('Churn Rate by Income Class')
plt.xlabel('Income Class')
plt.ylabel('Churn Rate')
plt.ylim(0, 1)  # Churn rate is between 0 and 1
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

**Explanation**

Calculate churn rate: We group the dataset by AnnualIncomeClass and calculate the mean of the Target variable, which represents the churn rate.

Reset Index: We reset the index of the resulting Series to get a DataFrame, which makes it easier to work with.

Grouped Bar Chart: We use plt.bar() to create a bar for each income class representing the churn rate.

Labels and Titles: We add appropriate titles and labels to make the chart informative.

**Result**

The x-axis will represent different income classes.

The y-axis will show the churn rate, which ranges from 0 to 1.

Each bar will reflect the proportion of customers who have churned in that income class.

This visualization will help you understand how churn rates differ across various income classes.

```python
# Correlation Heatmap: Use a heatmap to visualize the correlation between numeric variables (Age, Services Opted, Target).


# Step 1: Select numeric variables
numeric_vars = Customertravel[['Age', 'ServicesOpted', 'Target']]

# Step 2: Calculate the correlation matrix
correlation_matrix = numeric_vars.corr()

# Step 3: Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Step 4: Create the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True, linewidths=.5)

# Step 5: Add title
plt.title('Correlation Heatmap')
plt.show()
```

**Explanation**

Select Numeric Variables: We create a new DataFrame containing only the numeric variables you want to analyze.

Calculate Correlation Matrix: The .corr() function calculates the correlation coefficients between the selected numeric variables.

Heatmap Creation: We use seaborn's heatmap() function to visualize the correlation matrix:

* annot=True: Annotates each cell with the correlation coefficient.
* fmt=".2f": Formats the annotations to two decimal places.
* cmap='coolwarm': Sets the color palette for the heatmap.
* linewidths=.5: Adds lines between cells for clarity.

**Result**

The heatmap will display a matrix where:

* Each cell shows the correlation coefficient between the pairs of variables.
* Values close to 1 indicate a strong positive correlation, while values close to -1 indicate a strong negative correlation. Values around 0 indicate little to no correlation.


The color scale will help visualize the strength of the correlations.

This visualization can provide insights into how the selected numeric variables relate to one another.

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Step 1: Load your dataset
# Assuming customer_travel_data is your DataFrame and 'Feedback' is the column with customer reviews
# customer_travel_data = pd.read_csv('path_to_your_data.csv') # Uncomment and set the path if needed

# Step 2: Combine all feedback into a single string
feedback_text = ' '.join(Customertravel['FrequentFlyer'].dropna().astype(str))

# Step 3: Create a set of stopwords (optional)
stopwords = set(STOPWORDS)

# Step 4: Generate the word cloud
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      stopwords=stopwords,
                      min_font_size=10).generate(feedback_text)

# Step 5: Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off the axis
plt.title('Customer Feedback Word Cloud')
plt.show()
```

**Explanation**

Load the Dataset: The code assumes you have a DataFrame named customer_travel_data, with a column named 'FrequentFlyer' containing customer reviews. If needed, you can uncomment and set the correct path to load your dataset.

Combine Feedback: We concatenate all the feedback text into a single string using ' '.join(), while dropping any missing values (dropna()).

Stopwords: We create a set of stopwords using the predefined set from the wordcloud library. Stopwords are common words (like "and," "the," etc.) that we typically want to exclude from the word cloud.

Generate the Word Cloud: We create an instance of the WordCloud class, specifying the dimensions, background color, and the stopwords to exclude. The .generate() method creates the word cloud based on the combined feedback text.

Display the Word Cloud: Finally, we use matplotlib to visualize the word cloud with imshow(). The axis is turned off for a cleaner look, and we add a title.

**RESULT**

The resulting word cloud will display frequently mentioned words in larger fonts, indicating common themes or issues highlighted in customer FrequentFlyer. This visualization can provide quick insights into what customers are saying most often.
